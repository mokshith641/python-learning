```python
c_list=[]
n=int(input("size of the list"))
for i in range(n):
    c_list.append(int(input("enter the number {} to list".format(i))))
dup=[]
def duplicate(c_list):
    for i in range(len(c_list)):
        if c_list[i] not in dup:
            dup.append(c_list[i])
    print("a list before :",c_list)
    print("a list after:",dup)
duplicate(c_list)
```
